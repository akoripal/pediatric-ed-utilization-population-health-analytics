```python
healthcare_df = con.execute("""
SELECT
    ed.Geo_FIPS,
    ed.year,

    ed.Age0to17PopEst2015_19ACS AS pediatric_population,

    ed.NumberED_VisitsAge0to17,
    ed.NumberLowAcuityED_VisitsAge0to17,
    ed.NumberAsthmaRelatedED_Visits,
    ed.NumberOfInjuryRelatedVisits,
    ed.NumberAcuteRespiratoryTractInfectionRelatedED_VisitsAge0to17,

    pc.AllPrimaryCarePatientsAge0to17,
    pc.WellChildVisitsInPastYearAge0to17,
    pc.AsthmaDiagnosisAge0to17

FROM ed_visits ed
LEFT JOIN primary_care pc
ON ed.Geo_FIPS = pc.Geo_FIPS
AND ed.year = pc.year
""").fetchdf()
```

```python
healthcare_df.head()
```

```python
healthcare_df.shape
```

```python
healthcare_df["ed_visit_rate_per_1000"] = (
    healthcare_df["NumberED_VisitsAge0to17"] * 1000
    / healthcare_df["pediatric_population"]
)
```

```python
healthcare_df["asthma_ed_rate"] = (
    healthcare_df["NumberAsthmaRelatedED_Visits"] * 1000
    / healthcare_df["pediatric_population"]
)
```

```python
healthcare_df["injury_rate"] = (
    healthcare_df["NumberOfInjuryRelatedVisits"] * 1000
    / healthcare_df["pediatric_population"]
)
```

```python
healthcare_df["low_acuity_ratio"] = (
    healthcare_df["NumberLowAcuityED_VisitsAge0to17"]
    / healthcare_df["NumberED_VisitsAge0to17"]
)
```

```python
healthcare_df["primary_care_utilization"] = (
    healthcare_df["AllPrimaryCarePatientsAge0to17"]
    / healthcare_df["pediatric_population"]
)
```
